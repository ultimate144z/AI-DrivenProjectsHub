```python
# importing all necessary libraries . . .
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
import time
import pandas as pd
```

```python
# opening the web browser . . .
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
```

```python
# getting the url for most popular IMDB . . .
url = 'https://www.imdb.com/chart/moviemeter/?ref_=nv_mv_mpm'
    
# using driver.get() to open link in browser . . .
driver.get(url)

all_movie_details = []

num_movies = len(driver.find_elements(By.XPATH, "//ul[contains(@class, 'ipc-metadata-list')]/li//a[contains(@class, 'ipc-title-link-wrapper')]"))

for i in range(num_movies):
    try:
        movie_data = {}
        # Find the movie links again after navigating back . . .
        movie_links = driver.find_elements(By.XPATH, "//ul[contains(@class, 'ipc-metadata-list')]/li//a[contains(@class, 'ipc-title-link-wrapper')]")
        movie_url = movie_links[i].get_attribute('href')
        print(f"URL for movie {i+1}: {movie_url}")

        # Click the movie link to navigate to the movie's page . . .
        movie_links[i].click()
        time.sleep(3)
        
        # Scroll down to the bottom of the page
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait for the page to load
        time.sleep(5)
        
        driver.execute_script("window.scrollTo(0, 0);")  # Scrolls to the top of the page
            
        try:
            # Update the XPath to correctly target the movie title
            movie_title_element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CLASS_NAME, "hero__primary-text"))
            )
            movie_title = movie_title_element.text
            movie_data['Title'] = movie_title
        except TimeoutException:
            print(f"Could not find the title for movie {i+1}")
            
        try:
            # Update the XPath to correctly target the movie details
            movie_runtime_element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, "(//li[contains(@class, 'ipc-inline-list__item')])[7]"))
            )
            movie_runtime = movie_runtime_element.text
            movie_data['Runtime'] = movie_runtime
        except TimeoutException:
            print(f"Could not find the runtime for movie {i+1}")
            
        try:
            # Update the XPath to correctly target the movie details
            movie_rating_element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, "//span[contains(@class, 'sc-bde20123-1') and contains(@class, 'cMEQkK')]"))
            )
            movie_rating = movie_rating_element.text
            movie_data["Rating"] = movie_rating
        except TimeoutException:
            print(f"Could not find the rating for movie {i+1}")
    
        try:
            # Update the XPath to correctly target the movie details
            age_restriction_element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, "(//li[contains(@class, 'ipc-inline-list__item')])[6]"))
            )
            age_restriction = age_restriction_element.text
            movie_data["Age for movie"] = age_restriction
        except TimeoutException:
            print(f"Could not find the Age for movie {i+1}")
               
        try:
            # Update the XPath to correctly target the movie details
            movie_genre_element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, "//div[@class='ipc-chip-list--baseAlt ipc-chip-list']"))
            )
            movie_genre = movie_genre_element.text
            movie_data["Genre for movie"] = movie_genre
        except TimeoutException:
            print(f"Could not find the Genre for movie {i+1}")
                                               
                                               
        try:
            # Update the XPath to correctly target the movie details
            movie_writer_element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH,'//*[@id="__next"]/main/div/section[1]/div/section/div/div[1]/section[4]/ul/li[2]/div'))
            )
            movie_writer = movie_writer_element.text
            movie_data["Writers"] = movie_writer
        except TimeoutException:
            print(f"Could not find the writers for movie {i+1}")
    
        try:
            # Update the XPath to correctly target the movie details
            movie_director_element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, "//span[contains(text(), 'Director')]/following-sibling::div//ul"))
            )
            movie_director = movie_director_element.text
            movie_data["Director"] = movie_director
        except TimeoutException:
            print(f"Could not find the director for movie {i+1}")
                    
        all_movie_details.append(movie_data)
        
        # Going back to the movie list . . .
        
        driver.back()
        time.sleep(2)
        
        # Scroll down a bit to load more movies
        driver.execute_script("window.scrollBy(0, 100);")  # Adjust the value as needed for your page

        # Wait for the page to load the new movie elements
        time.sleep(2)

        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.XPATH, "//ul[contains(@class, 'ipc-metadata-list')]/li//a[contains(@class, 'ipc-title-link-wrapper')]"))
        )
        
    except Exception as e:
        print(f"Error: {e}")
        
data = pd.DataFrame(all_movie_details)

# Closing the WebDriver at the end of the scraping
# driver.quit()
```

```python
import pandas as pd
import numpy as np
import re
import random

# Sample data . . .
data = {
    "Title": ["The Killer", "The Marvels", "The Ballad of Songbirds & Snakes", "Tiger 3", "Madame Web", "Totally Killer", "Wingwomen", "The Wolf of Wall Street", "Dune", "Expend4bles"],
    "Runtime": ["1.0", "1.0", "2.0", "2h 34m", np.nan, "1.0", "1.0", "3.0", "2.0", "1.0"],
    "Rating": ["94K", np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan],
    "Age for movie": ["R", "PG-13", "PG-13", "2h 34m", "Claire Parker", "R", "TV-MA", "R", "PG-13", "R"],
    "Genre for movie": ["Action\nAdventure\nCrime", "Action\nAdventure\nFantasy", "Action\nAdventure\nDrama", "Action\nAdventure\nThriller", "Action\nAdventure\nSci-Fi", "Comedy\nHorror", "Action\nComedy\nCrime", "Biography\nComedy\nCrime", "Action\nAdventure\nDrama", "Action\nAdventure\nThriller"],
    "Writers": ["Alexis NolentLuc JacamonAndrew Kevin Walker", "Nia DaCostaMegan McDonnellElissa Karasik", "Michael LesslieMichael ArndtSuzanne Collins", "Shridhar RaghavanAnckur ChaudhryAditya Chopra", np.nan, np.nan, np.nan, "Terence WinterJordan Belfort", np.nan, np.nan],
    "Director": ["David Fincher", "Nia DaCosta", "Francis Lawrence", "Maneesh Sharma", "S.J. Clarkson", "Nahnatchka Khan", "Mélanie Laurent", "Martin Scorsese", "Denis Villeneuve", "Scott Waugh"]
}

# Creating DataFrame . . .
df = pd.DataFrame(data)

# Cleaning up the DataFrame . . .
df.fillna('Not Available', inplace=True)

# Converting 'Runtime' to minutes . . .
def convert_runtime_to_minutes(runtime):
    if 'h' in runtime:
        parts = runtime.split('h')
        hours = int(parts[0])
        minutes = int(parts[1].strip().replace('m', '')) if 'm' in parts[1] else 0
        return hours * 60 + minutes
    elif runtime.replace('.', '', 1).isdigit():
        return int(float(runtime) * 60)
    else:
        return 'Not Available'

df['Runtime'] = df['Runtime'].apply(convert_runtime_to_minutes)

# Split 'Genre for movie' . . .
df['Genre for movie'] = df['Genre for movie'].apply(lambda x: x.split('\n'))

# Ensuring all values in 'Writers' and 'Director' columns are strings . . .
df['Writers'] = df['Writers'].astype(str)
df['Director'] = df['Director'].astype(str)

# Function to split names in 'Writers' and 'Directors'  . . .
def split_names(names):
    if names == 'Not Available':
        return ['Not Available']
    return re.findall(r'[A-Z][^A-Z]*', names)

df['Writers'] = df['Writers'].apply(split_names)
df['Director'] = df['Director'].apply(split_names)

# Function to recommend a movie . . .
def recommend_movie(df, preferred_genre=None, max_runtime=None, age_restriction=None):
    filtered_df = df.copy()

    # Filter by Genre . . .
    if preferred_genre:
        filtered_df = filtered_df[filtered_df['Genre for movie'].apply(lambda genres: any(preferred_genre.lower() in genre.lower() for genre in genres))]

    # Filter by Maximum Runtime . . .
    if max_runtime is not None:
        filtered_df = filtered_df[filtered_df['Runtime'] != 'Not Available']
        filtered_df = filtered_df[filtered_df['Runtime'] <= max_runtime]

    # Filter by Age Restriction . . .
    if age_restriction:
        filtered_df = filtered_df[filtered_df['Age for movie'] == age_restriction]

    # If no movies match the criteria, recommend any movie from the original dataframe . . .
    if filtered_df.empty:
        filtered_df = df

    # Randomly select a movie to recommend . . .
    recommended_movie = filtered_df.sample(n=1).iloc[0]
    return recommended_movie

# Interactive part to get user preferences . . .
print("Welcome to the Movie Recommender!")
preferred_genre = input("Please enter your preferred genre (leave blank if no preference): ")
max_runtime_input = input("Please enter the maximum runtime in minutes (leave blank if no preference): ")
age_restriction = input("Please enter the age restriction (e.g., R, PG-13; leave blank if no preference): ")

# Convert max_runtime to integer if not blank . . .
max_runtime = int(float(max_runtime_input)) if max_runtime_input else None

# Get movie recommendation . . .
recommendation = recommend_movie(df, preferred_genre=preferred_genre, max_runtime=max_runtime, age_restriction=age_restriction)
print("\nMovie Recommendation:")
print(recommendation)
```
